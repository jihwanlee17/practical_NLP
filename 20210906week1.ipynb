```python
#lexical_diversity(text)함수를 만들자: 특정 텍스트에서 unique한 단어의 비율

def lexical_richness(text):
    return len(set(text))/len(text)
```

```python
lexical_richness(text5)
```

```python
#전체텍스트가 텍스트1에서 텍스트9까지인데 전체텍스트에 대해 lexical_richness결과 반환하기
#for loop

texts = [text1, text2, text3, text4, text5, text6, text7, text8, text9]

for text in texts:
    print(text, lexical_richness(text))
```
